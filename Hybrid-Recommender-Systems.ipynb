```python
data_recipy = rec[["id", "name"]]
data_recipy = data_recipy[data_recipy.id < 20000]
data_recipy.reset_index(drop=True, inplace=True)
data_recipy
```

```python
interactions = pd.read_csv("interactions_train.csv")
interactions_keep = interactions[["user_id", "recipe_id", "rating"]]
# interactions_keep[interactions_keep["recipe_id"]<5000]
interactions_keep = interactions_keep[interactions_keep.recipe_id < 20000]
interactions_keep.reset_index(drop=True, inplace=True)
interactions_keep
```

```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import ast 
from ast import literal_eval
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import linear_kernel, cosine_similarity
# from nltk.stem.snowball import SnowballStemmer
# from nltk.stem.wordnet import WordNetLemmatizer
# from nltk.corpus import wordnet
from surprise import SVD, Reader
from surprise import Dataset
from surprise.model_selection import cross_validate

import warnings; warnings.simplefilter('ignore')



reader = Reader()
data = Dataset.load_from_df(interactions_keep[['user_id', 'recipe_id', 'rating']], reader)

svd = SVD()
cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)

# trainset = data.build_full_trainset()
# svd.fit(trainset)
```

```python
data_recipy = data_recipy.dropna()
# data_recipy = data_recipy.loc[-50000:]
print(data_recipy.shape)
tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2), min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(data_recipy['name'])
tfidf_matrix.shape
```

```python


cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def hybrid(user_id, recipe_id, many):

    recipe_position_in_tfifd_matrix = data_recipy[data_recipy["id"]==recipe_id].index[0]

    sim_scores = list(enumerate(cosine_sim[int(recipe_position_in_tfifd_matrix)]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:100]

    recipe_indices = [i[0] for i in sim_scores]
    
    recipes = data_recipy.iloc[recipe_indices][['name', 'id']]
    
    recipes['est'] = recipes['id'].apply(lambda x: svd.predict(user_id, x).est)
    recipes = recipes.sort_values('est', ascending=False)
    return recipes.head(many)
```

```python
interactions_test = pd.read_csv("interactions_test.csv")
interactions_test_keep = interactions_test[["user_id", "recipe_id", "rating"]]
interactions_test_keep = interactions_test_keep[interactions_test_keep.recipe_id < 20000]
interactions_test_keep.reset_index(drop=True, inplace=True)
interactions_test_keep

```

```python
one = interactions_keep.loc[0]
# print(data_recipy[data_recipy["id"]==one["recipe_id"]])

recipes_user_rated = interactions_keep[interactions_keep["user_id"] == one.user_id]
print(recipes_user_rated)
all_recomendations = []
for recipe in recipes_user_rated.values:
    if recipe[-1]> 3:
        results = hybrid(one["user_id"], recipe[1], 10)
        all_recomendations.append(results)

all_df = pd.concat(all_recomendations)


recnames = [data_recipy[data_recipy["id"]==x[1]].name for x in recipes_user_rated.values]

print(recnames)

all_df = all_df.sort_values('est', ascending=False)
all_df
```

```python
merged = pd.merge(interactions_test_keep, interactions_keep, on="user_id", how="left", suffixes=["_test", "_train"])
merged = merged.dropna()
merged = merged.drop_duplicates(subset=['user_id'])
merged.reset_index(drop=True, inplace=True)
print(merged)
```
